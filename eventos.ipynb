```python
class Participante:
    documento = ""
    nombre = ""
    edad = ""
    cargo = ""
    pagado = False

    def mostrar():
        print(f"Documento: {Participante.documento}, Nombre: {Participante.nombre}, Edad: {Participante.edad}, Cargo: {Participante.cargo}, Pagado: {Participante.pagado}")

class Evento:
    nombre_evento = ""
    locacion = ""
    dia_mes = ""
    realizado = False

    def mostrar():
        print(f"Nombre del Evento: {Evento.nombre_evento}, Locación: {Evento.locacion}, Día del Mes: {Evento.dia_mes}, Realizado: {Evento.realizado}")

def registrar_participante(participantes):
    documento = input("Ingrese el documento del participante: ")
    nombre = input("Ingrese el nombre del participante: ")
    edad = input("Ingrese la edad del participante: ")
    cargo = input("Ingrese el cargo del participante: ")
    participante = Participante()
    participante.documento = documento
    participante.nombre = nombre
    participante.edad = edad
    participante.cargo = cargo
    participantes.append(participante)
    print("Participante registrado exitosamente.")

def quitar_participante(participantes, documento):
    for participante in participantes:
        if participante.documento == documento:
            if participante.pagado:
                print("No se puede eliminar este participante porque ya ha pagado.")
            else:
                participantes.remove(participante)
                print("Participante eliminado exitosamente.")
            return
    print("No se encontró ningún participante con ese documento.")

def mostrar_participantes(participantes):
    print("Listado de participantes:")
    for participante in participantes:
      participante.mostrar()

def calcular_deuda(participantes):
    deuda = 0
    for participante in participantes:
        if not participante.pagado:
            deuda += 50000
    return deuda

def listar_participantes_sin_pago(participantes):
    print("Participantes que no han pagado:")
    for participante in participantes:
        if not participante.pagado:
            participante.mostrar()

def registrar_pago(participantes):
    documento = input("Ingrese el documento del participante que va a pagar: ")
    for participante in participantes:
        if participante.documento == documento:
            participante.pagado = True
            print("Pago registrado exitosamente.")
            return
    print("No se encontró ningún participante con ese documento.")

def registrar_evento(eventos):
    nombre_evento = input("Ingrese el nombre del evento: ")
    locacion = input("Ingrese la locación del evento: ")
    dia_mes = input("Ingrese el día del mes del evento: ")
    evento = Evento()
    evento.nombre_evento = nombre_evento
    evento.locacion = locacion
    evento.dia_mes = dia_mes
    eventos.append(evento)
    print("Evento registrado exitosamente.")

def modificar_evento(eventos, nombre_evento):
    for evento in eventos:
        if evento.nombre_evento == nombre_evento:
            if evento.realizado:
                print("No se puede modificar un evento que ya ha sido realizado.")
                return
            evento.nombre_evento = input("Ingrese el nuevo nombre del evento: ")
            evento.locacion = input("Ingrese la nueva locación del evento: ")
            evento.dia_mes = input("Ingrese el nuevo día del mes del evento: ")
            print("Evento modificado exitosamente.")
            return
    print("No se encontró ningún evento con ese nombre.")

def eliminar_evento(eventos, nombre_evento):
    for evento in eventos:
        if evento.nombre_evento == nombre_evento:
            if evento.realizado:
                print("No se puede eliminar un evento que ya ha sido realizado.")
                return
            eventos.remove(evento)
            print("Evento eliminado exitosamente.")
            return
    print("No se encontró ningún evento con ese nombre.")

def marcar_evento_realizado(eventos, nombre_evento):
    for evento in eventos:
        if evento.nombre_evento == nombre_evento:
            evento.realizado = True
            print("Evento marcado como realizado exitosamente.")
            return
    print("No se encontró ningún evento con ese nombre.")

def mostrar_eventos(eventos):
    print("Listado de eventos:")
    for evento in eventos:
        evento.mostrar()

def main():
    participantes = []
    eventos = []

    while True:
        print("\n--- Menú Principal ---")
        print("1. Registrar participante")
        print("2. Quitar participante")
        print("3. Mostrar participantes")
        print("4. Registrar pago")
        print("5. Calcular deuda total")
        print("6. Listar participantes sin pago")
        print("7. Registrar evento")
        print("8. Modificar evento")
        print("9. Eliminar evento")
        print("10. Marcar evento como realizado")
        print("11. Mostrar eventos")
        print("12. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            registrar_participante(participantes)
        elif opcion == "2":
            documento = input("Ingrese el documento del participante que desea quitar: ")
            quitar_participante(participantes, documento)
        elif opcion == "3":
            mostrar_participantes(participantes)
        elif opcion == "4":
            registrar_pago(participantes)
        elif opcion == "5":
            print(f"La deuda total es: {calcular_deuda(participantes)} COP")
        elif opcion == "6":
            listar_participantes_sin_pago(participantes)
        elif opcion == "7":
            registrar_evento(eventos)
        elif opcion == "8":
            nombre_evento = input("Ingrese el nombre del evento que desea modificar: ")
            modificar_evento(eventos, nombre_evento)
        elif opcion == "9":
            nombre_evento = input("Ingrese el nombre del evento que desea eliminar: ")
            eliminar_evento(eventos, nombre_evento)
        elif opcion == "10":
            nombre_evento = input("Ingrese el nombre del evento que desea marcar como realizado: ")
            marcar_evento_realizado(eventos, nombre_evento)
        elif opcion == "11":
            mostrar_eventos(eventos)
        elif opcion == "12":
            confirmacion = input("¿Está seguro de que desea salir? (S/N): ").upper()
            if confirmacion == "S":
              print("Salio de eventos")
              break
        else:
          print("Opcion no valida, seleccione una opcion valida")

if __name__ == "__main__":
    main()
```
