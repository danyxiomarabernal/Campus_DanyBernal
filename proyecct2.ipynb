```python
class CampusLands:
    def __init__(self):
        self.campers = []
        self.rutas_entrenamiento = []

    def procesar_inscripcion(self, camper_data):
        nuevo_camper = Camper(**camper_data)
        self.campers.append(nuevo_camper)
        print(f"Inscripción procesada para {nuevo_camper.nombres} {nuevo_camper.apellidos}")

    def asignar_ruta_entrenamiento(self, camper_id, ruta_entrenamiento):
        camper = next((c for c in self.campers if c.id == camper_id), None)
        if camper:
            ruta = next((r for r in self.rutas_entrenamiento if r.nombre == ruta_entrenamiento), None)
            if ruta:
                if len(ruta.campers_asignados) < ruta.capacidad_maxima:
                    camper.ruta_entrenamiento = ruta.nombre
                    ruta.campers_asignados.append(camper)
                    print(f"Ruta de entrenamiento asignada a {camper.nombres} {camper.apellidos}")
                else:
                    print(f"No hay cupo disponible en la ruta de entrenamiento {ruta.nombre}")
            else:
                print(f"No se encontró la ruta de entrenamiento {ruta_entrenamiento}")
        else:
            print(f"No se encontró al Camper con ID {camper_id}")

    def registrar_evaluacion(self, camper_id, modulo, nota_teorica, nota_practica, nota_trabajos):
        camper = next((c for c in self.campers if c.id == camper_id), None)
        if camper:
            evaluacion = Evaluacion(camper, modulo, nota_teorica, nota_practica, nota_trabajos)
            camper.evaluaciones.append(evaluacion)
            print(f"Evaluación registrada para {camper.nombres} {camper.apellidos} en el módulo {modulo.nombre}")
        else:
            print(f"No se encontró al Camper con ID {camper_id}")

    def cambiar_estado_camper(self, camper_id, nuevo_estado):
        camper = next((c for c in self.campers if c.id == camper_id), None)
        if camper:
            camper.estado = nuevo_estado
            print(f"Estado de {camper.nombres} {camper.apellidos} cambiado a {nuevo_estado}")
        else:
            print(f"No se encontró al Camper con ID {camper_id}")

    def consultar_campers_en_riesgo_alto(self):
        campers_en_riesgo_alto = [camper for camper in self.campers if camper.riesgo == "Alto"]
        if campers_en_riesgo_alto:
            print("Campers en riesgo alto:")
            for camper in campers_en_riesgo_alto:
                print(f"- {camper.nombres} {camper.apellidos}")
        else:
            print("No hay campers en riesgo alto")


```
